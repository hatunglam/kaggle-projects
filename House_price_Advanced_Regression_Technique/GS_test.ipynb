```python
def split_feature(data):
    data_numeric = data.select_dtypes(['int', 'float']).drop(columns=['MSSubClass', 'OverallQual', 'OverallCond'], axis= 1)

    data_ordinal = data[['MSSubClass', 'OverallQual', 'OverallCond', 'GarageCond', 
                   'GarageQual', 'KitchenQual', 'HeatingQC', 'BsmtExposure', 
                   'BsmtCond', 'BsmtQual', 'ExterCond', 'ExterQual']]
    
    data_category = data.select_dtypes(['object']).drop(columns= ['GarageCond', 'GarageQual', 'KitchenQual',
                                                                            'HeatingQC', 'BsmtExposure', 'BsmtCond', 
                                                                            'BsmtQual', 'ExterCond', 'ExterQual'], axis= 1)
    
    return data_numeric.columns, data_ordinal.columns, data_category.columns
    
```

## 2. Data Cleaning and Preprocessing


```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder, OrdinalEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
data_numeric_cols, data_ordinal_cols, data_category_cols = split_feature(df2)

# Impute data

numeric_imputer = SimpleImputer(strategy='median')
ordinal_imputer = SimpleImputer(strategy= 'most_frequent')
categorical_imputer = SimpleImputer(strategy='most_frequent')

impute_df = ColumnTransformer(
    transformers=[
        ('num', numeric_imputer, data_numeric_cols),
        ('ord', ordinal_imputer, data_ordinal_cols),
        ('cat', categorical_imputer, data_category_cols)
    ],
    verbose_feature_names_out=False
    )
```

```python
##############################################################

imputed_data = impute_df.fit_transform(df2)
imputed_df2 = pd.DataFrame(imputed_data, columns=impute_df.get_feature_names_out(), index=df2.index)
imputed_df2.info()
```

```python
# Feature engineering

def feature_engineering(df):
    df['House_age'] = df['YrSold'] - df['YearBuilt']
    df.drop(columns=['YrSold', 'YearBuilt'], axis= 1, inplace= True)

    df.drop(columns=['1stFlrSF', '2ndFlrSF'], axis= 1, inplace= True)

    df['Total_Bath'] = df['BsmtFullBath'] + df['BsmtHalfBath'] + df['FullBath'] + df['HalfBath']
    df.drop(columns=['BsmtFullBath', 'BsmtHalfBath',
                     'FullBath', 'HalfBath'], axis= 1, inplace= True)
    
    df['bed_ratio'] = df['BedroomAbvGr'] / df['TotRmsAbvGrd'].replace({0: 1})

    df['sqft_per_car'] = df['GarageArea'] / df['GarageCars'].replace({0: 1})

    df['Total_Porch'] = df['OpenPorchSF'] + df['EnclosedPorch'] + df['3SsnPorch'] + df['ScreenPorch']
    df.drop(columns=['OpenPorchSF', 'EnclosedPorch', 
                     '3SsnPorch', 'ScreenPorch'], axis= 1, inplace= True)

    return df    
```

```python
df3 = feature_engineering(imputed_df2)
df3.head()
```

```python
#Encoding
ohe = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
ord_enc = OrdinalEncoder()

cat_encode = ColumnTransformer(
    transformers= [
        ('ordinal', ord_enc, data_ordinal_cols),
        ('one_hot', ohe, data_category_cols)
    ],
    remainder='passthrough',
    verbose_feature_names_out=False
)
```

```python
encoded_df = cat_encode.fit_transform(df3)
```

```python
feature_names = cat_encode.get_feature_names_out()
df_preprocessed = pd.DataFrame(encoded_df, columns=feature_names)
```
